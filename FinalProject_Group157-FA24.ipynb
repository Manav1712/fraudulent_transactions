## Background and Prior Work

Cryptocurrency is a type of digital currency that uses cryptography to secure digital data and assets. Blockchain is the underlying technology behind cryptocurrencies. It consists of a decentralized public ledger of all transactions that is updated and shared across many computers in a network. In 2015, the Ethereum blockchain was launched as a new platform, building on Bitcoin's success.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)
Ethereum blockchain is a decentralized blockchain platform that operates without control by any single entity, using its native digital currency, Ether. It supports various use cases, including financial services, games, social networks and other applications that anyone can build and use. More commonly, its blockchain technology facilitates the exchange of cryptocurrencies and digital transactions. Ethereum makes financial services more accessible to those without traditional payment providers and banks, and empowers users with greater control over their digital assets. 

However, Ethereum cryptocurrency could be exploited for fraudulent activities such as money laundering, Ponzi schemes, and other illicit financial behaviors.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Therefore, it is crucial to accurately detect anomalies and distinguish malicious transactions from legitimate ones. This is essential for mitigating potential risks, preventing malicious activities, and upholding data integrity. By maintaining the security and trustworthiness of the platform, we enhance the overall integrity of our cryptocurrency financial systems. 

Several studies have aimed to find the most robust and reliable anomaly detection systems in blockchain networks. One such study<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) utilizes machine learning tree-based ensemble classifiers combined with over-sampling and under-sampling techniques to detect anomalies on a dataset of Bitcoin transactions. Compared to other methods evaluated, this approach led to the highest true positive rate (TPR) and lowest false positive rate (FPR) values. This demonstrates the strategy's effectiveness in accurately detecting Blockchain anomalous transactions.

Another study<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) focused on detecting abnormal transactions specifically within the Ethereum network using a method known as ATD-SGAN, which employs a semi-supervised generative adversarial network. This was compared against existing solutions of intrusion detection systems (IDS) that monitor network traffic for signs of cyber-attacks or malicious activity. The results show that ATD-SGAN achieved a more refined classification of normal and abnormal transactions, thereby improving the average false negative and positive rates.

Another interesting study<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) applies machine learning techniques to detect anomalous transactions in a Blockchain dataset composed of Ethereum and Bitcoin network transactions data. A key component of this study was that it focuses on the assessment of the blockchain network communication and the reliability of the connection data, rather than ledger data. The results indicate higher accuracy and efficiency in its anomaly detection system compared to existing models of its time.

### Baseline Model: Logistic Regression
Logistic regression is a statistical model used to predict a binary outcome based on one or more independent variables. It models the probability of the outcome as a sigmoid (logistic) function of the predictors. In our case, we aim to predict the `flag` variable, where `1` indicates an illicit Ethereum address and `0` a non-illicit address.

**Pre-Processing Steps**

* **Scaling**: Features were standardized to have a mean of 0 and standard deviation of 1 using `StandardScaler`. This ensures that all features contribute equally to the model and avoids biases due to differing scales.
* **Variance Inflation Factor (VIF)**: VIF detects multicollinearity among features. High VIF values (>10) indicate redundancy, leading to numerical instability. We removed features with high VIF, simplifying the model and ensuring robust performance.

```python
# Prepare the features and target variable
features = ethereumData.drop(columns=['address', 'flag'])
target = ethereumData['flag']

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_features_df = pd.DataFrame(scaled_features, columns=features.columns)

X_train, X_test, y_train, y_test = train_test_split(scaled_features_df, target, test_size=0.3, random_state=42)
```

```python
vif_data = pd.DataFrame()
vif_data['Feature'] = X_train.columns
vif_data['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
```

```python
# Drop features with high VIF (>10)
high_vif_features = vif_data[vif_data['VIF'] > 10]
features_reduced = scaled_features_df.drop(columns=high_vif_features['Feature'])
# Display VIF information
print("VIF Data:")
print(vif_data)
print("High VIF Features:")
print(high_vif_features)
```

```python
# Re-split the data with reduced features
X_train_reduced, X_test_reduced, y_train, y_test = train_test_split(
    features_reduced, target, test_size=0.3, random_state=42
)
X_train_reduced_const = sm.add_constant(X_train_reduced)

# Fit the logistic regression model with reduced features
logit_model_reduced = sm.Logit(y_train, X_train_reduced_const)
result_reduced = logit_model_reduced.fit()

# Display model summary
print(result_reduced.summary())
```

**Interpretation of the Summary (Statsmodels)**
From the logistic regression summary:

* **Coefficients**: Positive coefficients (e.g., `minValSent`, `giniRec`) increase the probability of the target being 1, indicating stronger associations with illicit behavior. Negative coefficients (e.g., `activityDays`, `txFreq`) decrease this likelihood.
* **Significance**: Features with p-values < 0.05 are statistically significant predictors. Examples include:
    * `activityDays` (negative, highly significant): More active days reduce the likelihood of an address being illicit.
    * `minValSent` (positive, highly significant): Higher minimum value sent strongly correlates with illicit behavior.
    * `giniRec` (positive, significant): A higher Gini index for received values suggests concentration and links to illicit activity.

```python
# Predict probabilities on training and testing sets
train_pred_probs = result_reduced.predict(X_train_reduced_const)
test_pred_probs = result_reduced.predict(sm.add_constant(X_test_reduced))

# Convert probabilities to binary predictions (default threshold = 0.5)
train_preds = (train_pred_probs >= 0.5).astype(int)
test_preds = (test_pred_probs >= 0.5).astype(int)
```

```python
# Calculate accuracy for train and test datasets
train_accuracy = accuracy_score(y_train, train_preds)
test_accuracy = accuracy_score(y_test, test_preds)

# Compute confusion matrices for train and test datasets
train_conf_matrix = confusion_matrix(y_train, train_preds)
test_conf_matrix = confusion_matrix(y_test, test_preds)

plt.figure(figsize=(12, 6))

# Training confusion matrix
plt.subplot(1, 2, 1)
sns.heatmap(train_conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Confusion Matrix - Training Data")
plt.xlabel("Predicted")
plt.ylabel("Actual")
```

```python
# Testing confusion matrix
plt.subplot(1, 2, 2)
sns.heatmap(test_conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Confusion Matrix - Testing Data")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.tight_layout()
plt.show()
```

```python
print(f'Test Accuracy of Logistic Regression Model {test_accuracy}')
```

```python
# Calculate precision, recall, F1 score, and AUC for training and testing sets
train_precision = precision_score(y_train, train_preds)
train_recall = recall_score(y_train, train_preds)
train_f1 = f1_score(y_train, train_preds)
train_auc = roc_auc_score(y_train, train_pred_probs)

test_precision = precision_score(y_test, test_preds)
test_recall = recall_score(y_test, test_preds)
test_f1 = f1_score(y_test, test_preds)
test_auc = roc_auc_score(y_test, test_pred_probs)
```

### Alternate Model: XGBoost
XGBoost, or Extreme Gradient Boosting, is a powerful machine learning algorithm primarily used for classification. It is based on the Decision Tree architecture. It builds decision trees sequentially (boosting), where each tree attempts to correct the errors of the previous one.  XGBoost is particularly well-suited for high-dimensional and tabular datasets because of the following reasons:

#### Why XGBoost is Effective for Anomaly Detection

1. **Handles High-Dimensional Data**  
   XGBoost performs well with datasets that have many features, enabling it to identify patterns that may indicate anomalies.

2. **Captures Non-Linear Relationships**  
   Its tree-based structure excels at modeling non-linear interactions, which are often critical in distinguishing anomalies from normal data.

3. **Robust to Imbalanced Data**  
   XGBoost can handle imbalanced datasets effectively by assigning higher weights to rare anomaly samples, ensuring they are not overlooked.

4. **Interpretable Results**  
   Feature importance values provide insights into why certain points are flagged as anomalies.

While picking this model, we could have also picked another popular machine learning algorithm - Random Forests. These use a technique called Bagging, which is essentially running a bunch of decision trees simultaneously and averaging the results. However, we picked XGBoost because it uses boosting, making it effective for modeling temporal (time based) anomalies that may evolve over time.


```python
# Initialize the XGBoost classifier
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)

# Train the model on the training data
xgb_model.fit(X_train_reduced, y_train)
```

## A Different and more Complex Approach - Using an Autoencoder

Autoencoders are neural networks designed to learn a compressed representation of input data by reconstructing it as closely as possible. Neural networks, in general, are computational models inspired by the human brain, consisting of layers of interconnected nodes (neurons) that process and learn patterns in data through training. They are widely used in tasks such as classification, regression, and anomaly detection due to their ability to capture complex relationships in high-dimensional data. Here's why they could work well for this task:

1. **Reconstruction Error as a Metric**:
   - Autoencoders learn to minimize the reconstruction error (Mean Squared Error) for data similar to the training set.
   - Anomalies / Illicit Transactions, which differ significantly from the normal patterns, typically result in higher reconstruction errors.

2. **Flexibility**:
   - They can be applied to diverse data types, such as tabular data, time series, and images.

4. **Dimensionality Reduction**:
   - By mapping inputs to a lower-dimensional latent space, autoencoders focus on essential features, reducing noise and making it easier to identify deviations.

In this case, the autoencoder will learn the common patterns in the transactional and financial metrics of Ethereum addresses. 

#### Why Use an Autoencoder for Anomaly Detection?

1. **No Need for Labels**  
   Autoencoders are unsupervised, meaning they don't require labeled data. This is useful for anomaly detection when labels for anomalies are sparse or unavailable, unlike logistic regression or XGBoost which rely on labeled training data. We also wanted to try a completely different method!

2. **Captures Complex Patterns**  
   Autoencoders can model nonlinear relationships and subtle patterns in high-dimensional data, something simpler models like logistic regression might miss.

4. **Better for Unknown Anomalies**  
   In real-world scenarios where the characteristics of anomalies are unknown or constantly changing, autoencoders adapt well without needing predefined rules or labels.


```python
#Feature scaling technique that transforms the data into a specific range
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_train_reduced_const)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_train, test_size=0.2, random_state=42)
```

```python
input_dim = X_train.shape[1]

# Building the Autoencoder architecture
input_layer = Input(shape=(input_dim,))
encoded = Dense(64, activation='relu')(input_layer)
encoded = Dense(32, activation='relu')(encoded)

decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(input_dim, activation='sigmoid')(decoded)

autoencoder = Model(input_layer, decoded)

# Compile the autoencoder model with Adam optimizer and mean squared error loss function
autoencoder.compile(optimizer='adam', loss='mse')


history = autoencoder.fit(
    X_train, X_train,
    epochs=50,
    batch_size=32,
    shuffle=True,
    validation_split=0.2
)
```

```python
train_reconstructions = autoencoder.predict(X_train)
test_reconstructions = autoencoder.predict(X_test)
train_recon_errors = np.mean(np.square(X_train - train_reconstructions), axis=1)
test_recon_errors = np.mean(np.square(X_test - test_reconstructions), axis=1)
```

```python
threshold = np.percentile(train_recon_errors, 95)
print(f"Anomaly threshold: {threshold}")
```
